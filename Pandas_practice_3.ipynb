# Creating new columns

Lets create a new column from a series

```python
car_sales
```

```python
car_sales["Seats"]=pd.Series([5,5,5,5,5,5])
car_sales
```

```python
car_sales["Seats"].fillna(5,inplace=True)
```

```python
car_sales
```

```python
car_sales["Utilization"]=car_sales["Odometer (KM)"]/car_sales["Price"]
car_sales
```

```python
car_sales.groupby(["Make"]).mean()
```

```python
car_sales.groupby(["Utilization"]).mean()
```

Lets create another column using python list

```python
fuel_used = [12.4,23,5,6,8.2,9.2,10,11,9,12]
car_sales["Fuel per 100 KM"]=fuel_used
car_sales
```

```python
car_sales["Total Fuel Used"]=car_sales["Odometer (KM)"]/100*car_sales["Fuel per 100 KM"]
car_sales
```
